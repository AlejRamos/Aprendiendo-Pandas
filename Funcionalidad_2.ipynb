```python
import pandas as pd
import numpy as np

"""
Estadísticas Descriptivas
"""
```

```python
df = pd.DataFrame(
    {
        "one": pd.Series(np.random.randn(3), index=["a", "b", "c"]),
        "two": pd.Series(np.random.randn(4), index=["a", "b", "c", "d"]),
        "three": pd.Series(np.random.randn(3), index=["b", "c", "d"]),
    }
)

df
```

```python
"""
Existen un monton de metodos para calcular estadísticas descripticas y otras
operaciones relacionadas en Series y DataFrame. Muchos de estos son agregaciones
es decir, el resultado es de una dimension, tales como sum(), mean() y 
quantile(), pero hay otros como cumsum() y cumprod(), que producen un objeto
del mismo tamaño.
Estos métodos toman un axis como argumento, pero los axis pueden ser
especificados por nombre o entero.
Para Series: No se necesitan axis
DataFrame: "index" (axis = 0, por default), "columns" (axis=1)
"""

print(df)
print()
#Para obtener el promedio de los indices de un DataFrame
print(df.mean(0))
print()
#Promedio de las columnas
print(df.mean(1))
print()
#Suma de columna o index
print(df.sum(1))
print()
#Quantile, Devuelve los valores en el cuantil dado sobre el eje solicitado
print(df.quantile())
print()
# Suma acumulativa
print(df.cumsum()) #Preserva la locacion de los NaN
print()
#Promedio acumulativo
print(df.cumprod()) #Preserva la locacion de NaN
```

```python
"""
Método usando skipna
Como su nombre intruye, skipna se utiliza para excluir los datos faltantes
NaN
"""

#El promedio usando skipna=False, por default True
print("\nUsando skipna=False no se se excluyen NaN\n",
      df.mean(axis=0, skipna=False))
#Ahora usando skipna = True
print("\nUsando skpna=True, se excluyen los NaN\n",
      df.mean(axis=0, skipna=True))
```

```python
"""
Combinando el comportamiento transmision/aritmetica podemos describir
varios procesos aritmeticos, tales como la estandarización
"""
#Es el método std() el encargado de estandarizar
print("La estandarización de columas es\n",
      df.std(1))
#Esta es la forma de hacer unitarip
ts_stand = (df - df.mean()) / df.std()

print(ts_stand.std())

print("\nEn el link\nhttps://pandas.pydata.org/docs/user_guide/basics.html#descriptive-statistics\npodemos encontrar una tabla con más opciones y su descripción")
```

```python
"""
Ultimas anotaciones de esta primer parte
"""
print("Nota: EN los cambios de métodos Numpy, como mean, std, y sum\n",
      "se excluyen por default Nas en las entradas de Series\n\n")

#Obtenemos el promedio usando método numpy
print("El promedio de la columna one usando np.mean() es:\n",
      np.mean(df["one"]))

#Antes de obtener el promedio, haremos la columna one un objeto numpy
print("Primero hacemos un objeto numpy a la columna one\n",
      df["one"].to_numpy(),
      "\nLuego obtenemos el promedio\n",
      np.mean(df["one"].to_numpy()))

"""
El metodo Series.nunique() retorna el noumpero de valores no-NaN
únicos en una Serie
"""
#Creamos una serie
series =pd.Series(np.random.randn(500))
#Hacemos NaN algunos valores
series[20:500] = np.nan
#Hacemos valores no-NaN 
series[15:20] = 5

print("\n\nAplicamos el método series.unique()\n",
      "La cantidad de valores no-NaN únicos son\n",
      series.nunique())
```

```python
"""
Resumen de los datos: describe
Existe una función conveniente, conocida como describe() la cual
arroja un resumen estdístico de los datos, exluyendo por supuesto 
NaN.
"""

#Creamos una serie de número random
series1 = pd.Series(np.random.rand(500))
#El siguiente comando indica que, cada 4 elementos agregué un NaN
series1[::4] = np.nan

#Imprimimos el resumen de la Serie
print("La Serie contiene:\n",
      series1.describe())

#Por supuesto funciona también con DataFrame

frame = pd.DataFrame(np.random.randn(200, 5), 
                     columns=["a", "b", "c", "d", "e"])

frame[::5] = np.nan

#Imprimimos el resumen
print("\nEl DataFrame contiene\n",frame.describe(),
      "\n\nObserve que para DataFrame el resumen es por columna")
print("\n\nObservese que existe una cantidad de porcentajes,",
      "\nestos se pueden elegir a conveniencia")
print("\nLa serie contiene la siguiente información\n"
    ,series1.describe(percentiles=[0.10,0.50,0.80]))
```

```python
"""
describe, segunda parte
Para objetos no númericos, describe() nos otorga una resumen simple
del número único de de valores, y los más frecuentes.
"""

#Creamos una serie no númerica
s = pd.Series(["a", "a", "b", "b", "a", "a", np.nan, "c", "d", "a"])

#Imprimimos el resumen
print("El resumen de la serie s es\n",s.describe())

"""
Por otro lado, para objetos mixtos, describe() se restringe
a dar un resumen de columnas númericas, y en caso de no contar con ninguna
solo columnas categoricas 
"""

#Cramos un DataFrame mixto
frame = pd.DataFrame({"a": ["Yes", "Yes", "No", "No"], "b": range(4)})

#Imprimimos el resumen
print("\nEl resumen de un DataFrame mixto es:\n",
      frame.describe())

"""
Pero también podemos controlar el comportamiento en el caso de objetos
mixtos, a través de los argumentos include/exclude. También se puede
usar all.
"""

#Imprimamos el resumen de la columna no númerica
print("\nEL resumen del DataFrame de la columna no númerica es:\n",
      frame.describe(include=['object']))

#Imprimimos el resumen de la columa numérica
print("\nEl resumen del DataFrame de la columna númerica es:\n",
      frame.describe(include=["number"]))

#Imprimos el resumen de todo
print("El resumen total es:\n",
      frame.describe(include="all"))
```

```python
"""
Index de los valores min/max
Las funciones idxmin() y idxmax() en Series y DataFrame devuelve la etiqueta
del index del mínimo y máximo
"""

#Creamos una Serie
s1 = pd.Series(np.random.randn(5))

print("La serie\n",
      s1,
      "\nContiene sus valores mínimo y máximo respectivamente en:\n",
      (s1.idxmin(),s1.idxmax()))

#En el caso de un DataFrame lo hace por columnas y nos da la columna y el
#Label del index

#Creamos un DataFrame
df1 = pd.DataFrame(np.random.randn(5, 3), columns=["A", "B", "C"])

print("\nEl DataFrame\n",
      df1,
      "\nTiene sus valores mínimos en:\n",
      df1.idxmin(),
      "\ny sus valores máximos en:\n",
      df1.idxmax())

#Como se puede notar por default cálcula el valor en las columnas
#Se puede hacer por filas

print("\nLos valores máximos y mínimos por filas son:\n",
      df1.idxmax(axis=1),
      df1.idxmin(axis=1))
print("\n\nNOTA IMPORTANTE:\nEn caso de contener repetidas veces el mismo valor minimo o máximo, devuelve la posición del primero en aparecer")
```

```python
"""
Value counts (histogramming) / mode
El método value_counts() de Series y top-level calcula un histograma de
valores en array 1-D.
"""

#Creamos un array
data = np.random.randint(0, 7, size=50)

#Hacemos una Serie
s = pd.Series(data)
print("Usando el método value_counts se obtiene:\n",
      s.value_counts(),
      "\nNótese que se ordenan de mayor a menor conteo\n")

"""
El método también puede usarse en una combinación de columnas. Por default
todas las columnas son usadas pero un subset puede ser seleccionado usando
el argumento select
"""

data1 = {"a": [1, 2, 3, 4], "b": ["x", "x", "y", "y"]}

frame = pd.DataFrame(data1)

print("El conteo de un DataFrame de una combinación de columnas es:\n",
      frame.value_counts())

"""
Por supuesto, también se puede encontrar el valor más frecuente, es decir
la moda en una Serie o DataFrame
"""

#Creamos una serie con claros valores repetidos
s5 = pd.Series([1, 1, 3, 3, 3, 5, 5, 7, 7, 7])

print("\nLa moda de la Serie\n",
      s5,
      "\nes:\n",
      s5.mode(),
      "\nDado que tenemos dos valores cuya frecuencia es igual hay dos modas")

#Es turno de un DataFrame
df5 = pd.DataFrame(
    {
        "A": np.random.randint(0, 7, size=50),
        "B": np.random.randint(-10, 15, size=50),
    }
)

print("\n\nEL DataFrame:\n",
      df5.head(),
      "\nTiene por moda:\n",
      df5.mode())
```

```python
"""
Discretización y Cuantización
Los valores continuos pueden ser discretizados con las funciones
cut() (bins basados en valores) y qcut() (bins basados en muestras de quantiles)

Utilice cut() cuando necesite segmentar y ordenar los valores de los datos
en contenedores. Esta función también es útil para pasar de una variable
continua a una variable categórica. Por ejemplo, el corte podría convertir 
las edades en grupos de edades. Admite el agrupamiento en un número igual 
de contenedores, o un conjunto predeterminado de contenedores.

qcut() Discretiza variable en cubos de igual tamaño basado en rango 
o basado en cuantiles de muestra. Por ejemplo, 1000 valores para 10 
cuantiles producirían un objeto categórico que indicaría la pertenencia 
a cuantiles para cada punto de datos.
"""

#
arr = np.random.randn(20)

factor = pd.cut(arr, 4)

print(factor)

factor = pd.cut(arr, [-5, -1, 0, 1, 5])

print(factor)

#calcula cuantiles de muestra. Por ejemplo, podríamos dividir algunos 
#datos normalmente distribuidos en cuartiles de igual tamaño como:"

arr = np.random.randn(30)

factor = pd.qcut(arr, [0, 0.25, 0.5, 0.75, 1])

print(factor,
      pd.value_counts(factor))

```
